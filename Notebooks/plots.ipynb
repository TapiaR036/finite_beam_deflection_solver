```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 3, 100)
y1 = 1.25*x**2 - 3.75*x


plt.plot(x, y1, label = '$ d^2y/dx^2 = 1.25x^2 - 3.75x$', color='blue')
plt.title("PDE Beam Deflection Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np

L = 10
x = np.linspace(0, L, 100)

plt.figure(figsize=(8,2))

# Beam line
plt.plot(x, np.zeros_like(x), linewidth=4)

# Supports
plt.scatter([0, L], [0,0], s=200)

# Distributed load arrows
for i in np.linspace(1,9,8):
    plt.arrow(i,0.5,0,-0.4,head_width=0.2)

plt.text(L/2,0.8,"Distributed Load q(x)", ha='center')
plt.text(0,-0.4,"Pinned Support")
plt.text(L,-0.4,"Pinned Support", ha='right')

plt.title("Beam Model Used for Numerical Simulation")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import math

nlog = [math.log10(10), math.log10(20), math.log10(50), math.log10(100)]
norm = [math.log10(0.0101603), math.log10(0.00227957), math.log10(0.000340377), math.log10(7.40612e-05)]
ref = [-2*math.log10(10), -2*math.log10(20), -2*math.log10(50), -2*math.log10(100)]

plt.plot(nlog, norm, label = '$l2 error$', color='blue')
plt.plot(nlog, ref, label= '$Reference line$', color='red')
plt.title("Convergence Rates (log format)")
plt.xlabel("log(n)")
plt.ylabel("log(l2)")
plt.legend()
plt.grid()
```
